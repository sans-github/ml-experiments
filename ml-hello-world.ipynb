```python
import openai
import traceback

openai.api_key = os.getenv('OPENAI_API_KEY')

def do_this(prompt, show_result=True):
    result = None
    try:
        completion = openai.Completion.create(
            engine="text-davinci-003",
            prompt="\n# ".join(prompt.split("\n")),
            echo=True,
            temperature=0,
            max_tokens=1024,
            top_p=0.1,
            frequency_penalty=0,
            presence_penalty=0
        )
        result = completion.choices[0].text
        if show_result:
            print(result)
        exec(result)
    except Exception as e:
        print("Program error : ", str(e))
        print(result)
    return result
```

```python
prompt = f"""
Build a model to predict Value based on Rooms and Distance. Use prices.csv file to train the model. 
Print top 5 rows from the training data
Show a 3D scatter plot of the model inline
Print the results for (Rooms, Distance)=(15, 61)
"""

result = do_this(prompt)
```
